## Rehshape with -1

```python
#resape if we dont know the number of items using -1
my_torch2=torch.arange(15)
my_torch2
```

```python
my_torch2=my_torch2.reshape(3,-1) # assuming we dont know that the value is 15 how do we determine how many rows to columns to create so we use -1
my_torch2
```

## Reshape v View
The main difference is that view() requires the tensor's underlying memory to be contiguous, while reshape() is more flexible and will automatically create a copy if the memory is not contiguous. 

```python
my_torch3=torch.arange(10)
my_torch3
```

```python
my_torch4=my_torch.view(2,5)
my_torch4
```

```python
# Reshape and view will update 
my_torch5=torch.arange(10)
my_torch5

my_torch6=my_torch5.reshape(2,5)
my_torch6

my_torch5[1]=4141
my_torch5
```

```python
my_torch6
#as you can see changing value is torch5 also updates torch 6
```
